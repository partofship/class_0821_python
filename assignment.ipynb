김다솜 강사님 라이브세션에서 나왔던 과제

```python
word = "Python"

# 예상 결과: h
# 예상 결과: n

# 여기에 코드 작성
print(word[-3])
print(word[-1])
```

```python
numbers = [10, 20, 30, 40, 50]
# 예상 결과: 30
# 예상 결과: 10

# 여기에 코드 작성
print(numbers[-3])
print(numbers[0])
```

```python
text = "Python"
# 예상 결과: Pyt
# 예상 결과: thon


# 여기에 코드 작성
print(text[:2])
print(text[2:])
```

```python
data = [10, 20, 30, 40, 50]
# 예상 결과: [20, 30, 40]
# 예상 결과: [50, 40, 30, 20, 10]

# 여기에 코드 작성
print(data[1:3])
print(data[::-1])
```
